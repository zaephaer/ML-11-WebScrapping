## Web Scraping using Beautiful Soup

Objective: Web Scraping using Beautiful Soup  
Ref: 

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
```

```python
url = "https://www.mudah.my/Johor/Properties-for-sale-2001?lst=0&fs=1&w=112&cg=2001&so=1&st=s&ps=0&pe=2"
uClient = uReq(url)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html, "html.parser")
type(page_soup)
```

## Create Container from Parsed Webpage

```python
containers = page_soup.findAll("div", {"class":"listing_params_container "}) #sc-iUVpnZ hsUkkn
#calculate number of item in container
print("Number of items in the container :", len(containers))
```

## Data Extraction

```python
container = containers[0]
print(soup.prettify(container))
```

```python
propertyName = container.find("h2", {"class":"list_title"}).text.strip()
print(propertyName)
```

```python
propertyPricex = container.find("div", {"class":"ads_price"}).text.strip().replace(" ","")
propertyPrice = propertyPricex.replace("RM","")
print(int(propertyPrice))
```

```python
propertyCategory = container.find("div", {"class":"display-block"}).text.strip()
print(propertyCategory)
```

```python
#propertySize = container.find("div", {"class":"bottom_info bottom_label size house-size"}).text
propertySize = container.find("div", attrs = {"title":"Size"}).text.strip()
print(propertySize)
```

```python
location = container.find("div", {"class":"area"}).text.strip()
print(location)
```

## Declare Extracted Data Column

```python
for container in containers:
    Name = container.find("h2", {"class":"list_title"}).text.strip()
    
    propertyPricex = container.find("div", {"class":"ads_price"}).text.strip().replace(" ","")
    Price = propertyPricex.replace("RM","")
    
    Category = container.find("div", {"class":"display-block"}).text.strip()
    Size = container.find("div", attrs = {"title":"Size"}).text.strip()
    #Bedroom = container.find("div", {"class":"bottom_info bottom_label bedroom "}).text.strip()
    #Bathroom = container.find("div", {"class":"bottom_info bottom_label bathroom "}).text.strip()
    Location = container.find("div", {"class":"area"}).text.strip()
    
    print("Property:" +Name)
    print("Price   :" +Price) 
    print("Category:" +Category)
    print("Size    :" +Size)
    #print("Bedroom :" +Bedroom)
    #print("Bathroom:" +Bathroom)
    print("Location:" +Location)
```

## Using Loop to Put Data in Dataframe

```python
dfNames = []
dfPrice = []
dfCategory = []
dfSize = []
dfLocation = []
```

```python
for container in containers:
    Name = container.find("h2", {"class":"list_title"}).text.strip()
    dfNames.append(Name)
    
    propertyPricex = container.find("div", {"class":"ads_price"}).text.strip().replace(" ","")
    Price = propertyPricex.replace("RM","")
    dfPrice.append(Price)
    
    Category = container.find("div", {"class":"display-block"}).text.strip()
    dfCategory.append(Category)
    
    Size = container.find("div", attrs = {"title":"Size"}).text.strip()
    dfSize.append(Size)
    
    Location = container.find("div", {"class":"area"}).text.strip()
    dfLocation.append(Location)   
```

## Create Dataframe

```python
import pandas as pd
df = pd.DataFrame({'Property': dfNames,
                   'Price': dfPrice, 
                   'Category': dfCategory, 
                   'Size': dfSize,
                   'Location': dfLocation
                  })
print(df.info())
```
