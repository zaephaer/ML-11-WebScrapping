## Convert to Dataframe

```python
content_list = []

#url = base_url+str(page_number)+"/.html"
r = requests.get(base_url)
c = r.text
soup = BeautifulSoup(c,"html.parser")
    
containers = soup.find("table",id="DataList1").find_all("td", class_="auto-style156")
    
for container in containers:
        content_dict = {}
        content_dict["var_1"] = container.find_all("tr",class_="auto-style136")[0].find_all("strong")[1].text.strip() 
        content_dict["var_2"] = container.find_all("tr",class_="auto-style136")[1].find_all("span")[0].text    
        content_dict["var_3"] = container.find_all("tr",class_="auto-style136")[1].find_all("span")[1].text
        content_dict["var_4"] = container.find_all("tr",class_="auto-style136")[1].find_all("span")[2].text 
        content_dict["var_5"] = container.find_all("tr",class_="auto-style136")[2].find("span").text
        content_dict["var_6"] = container.find_all("tr",class_="auto-style136")[3].find("span").text 
        content_dict["var_7"] = container.find_all("tr",class_="auto-style136")[4].find("span").text 
        content_dict["var_8"] = container.find_all("tr",class_="auto-style136")[5].find("span").text
        content_dict["var_9"] = container.find_all("tr",class_="auto-style136")[7].find("span").text 
        
        # To store the dictionary to into a list
        content_list.append(content_dict)
    
# To make a dataframe with the list
df = pd.DataFrame(content_list)
#print('At least {} pages scrapped, containing {} properties.' .format(last_page, len(var_1)))
```
