## Import Library / Make URL Request / Test Communication

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://www.auctionmart.my/search/?kw=&st=&pt=&lp=&hp=&ll=&hl=&ad=&ab=&tn="
r = requests.get(base_url)
c = r.text

print(r)
```

## Parse HTML

```python
# To parse the html
soup = BeautifulSoup(c,"html.parser")
```

## Calculate Number of item in Container

```python
containers = soup.find_all("div",class_="info_property_l")
print("Number of items in the container :", len(containers))
```

## Convert to Dataframe

```python
content_list = []

start_page = 1
for page_number in range(int(start_page),20):
    url = base_url+str(page_number)+"/.html"
    r = requests.get(base_url)
    c = r.text
    soup = BeautifulSoup(c,"html.parser")
    
    containers = soup.find_all("div",class_="info_property_l")
    
    for container in containers:
        content_dict = {}
        content_dict["var_1"] = container.find_all("li")[0].text
        content_dict["var_2"] = container.find_all("li")[1].text   
        content_dict["var_3"] = container.find_all("li")[2].text
        content_dict["var_4"] = container.find_all("li")[3].text 
        content_dict["var_5"] = container.find_all("li")[4].text
       
        # To store the dictionary to into a list
        content_list.append(content_dict)
    
# To make a dataframe with the list
df = pd.DataFrame(content_list)
#print('At least {} pages scrapped, containing {} properties.' .format(last_page, len(var_1)))
```
